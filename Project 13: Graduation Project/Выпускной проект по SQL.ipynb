1. **Посчитаем, сколько книг вышло после 1 января 2000 года**

```python
query = '''
           SELECT COUNT(*) 
           FROM books 
           WHERE publication_date > '2000-01-01';
'''
con=engine.connect()
pd.io.sql.read_sql(sql=text(query), con = con)
```

- После 1 января 2000 года вышло 819 книг

2. **Для каждой книги посчитаем количество обзоров и среднюю оценку**

```python
query = '''
           SELECT books.title, 
           COUNT(DISTINCT reviews.review_id) AS count_reviews, 
           AVG(ratings.rating) AS avg_rating
           FROM books
           LEFT JOIN reviews ON books.book_id = reviews.book_id
           LEFT JOIN ratings ON books.book_id = ratings.book_id
           GROUP BY books.book_id;
'''
con=engine.connect()
pd.io.sql.read_sql(sql=text(query), con = con)
```

- count_reviews - количество обзоров по каждой книге.
- avg_rating - средняя оценка по каждой книге.

3. **Определим издательство, которое выпустило наибольшее число книг толще 50 страниц — так исключим из анализа брошюры**

```python
query = '''
           SELECT publishers.publisher, 
           COUNT(books.book_id) AS num_books
           FROM books
           JOIN publishers ON books.publisher_id = publishers.publisher_id
           WHERE books.num_pages > 50
           GROUP BY publishers.publisher_id
           ORDER BY num_books DESC
           LIMIT 1;
'''
con=engine.connect()
pd.io.sql.read_sql(sql=text(query), con = con)
```

- Penguin Books - издательство, которое выпустило наибольшее число книг толще 50 страниц. 
- Penguin Books Выпустило 42 книги.

4. **Определим автора с самой высокой средней оценкой книг — учитываем только книги с 50 и более оценками**

```python
query = '''
           SELECT authors.author, AVG(ratings.rating) AS avg_rating
           FROM authors   
           JOIN books ON authors.author_id = books.author_id   
           JOIN ratings ON books.book_id = ratings.book_id   
           WHERE books.book_id IN   
           (SELECT book_id
           FROM ratings
           GROUP BY book_id
           HAVING COUNT(*) >= 50)
           GROUP BY authors.author_id
           ORDER BY avg_rating DESC
           LIMIT 1;
'''
con=engine.connect()
pd.io.sql.read_sql(sql=text(query), con = con)
```

- J.K. Rowling/Mary GrandPré - автор с самой высокой средней оценкой книг.

5. **Посчитаем среднее количество обзоров от пользователей, которые поставили больше 48 оценок**

```python
query = '''
           SELECT AVG(subquery.review_count) AS avg_review_count
FROM (
  SELECT COUNT(*) AS review_count
  FROM reviews
  WHERE username IN (
    SELECT username
    FROM (
      SELECT username, COUNT(*) AS rating_count
      FROM ratings
      GROUP BY username
      HAVING COUNT(*) > 48
    ) AS rating_subquery
  )
  GROUP BY username
) AS subquery;

'''
con=engine.connect()
pd.io.sql.read_sql(sql=text(query), con = con)
```

- Пользователи, которые поставили больше 48 оценок в среднем сделали 24 обзора.
